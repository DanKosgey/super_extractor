```python
corr_threshold = 0.8
corr_matrix = df_training_numeric.corr()
high_corr_features = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > corr_threshold:
            high_corr_features.add(corr_matrix.columns[i])

# Get the list of non-correlated features
selected_features = [col for col in df_training_numeric.columns if col not in high_corr_features]
print(f"Selected features: {selected_features}")
```
